*Import data from https://www.icphusa.org/wp-content/uploads/2016/04/Appendix.pdf*
* Turn pdf into excel file
* From excel file only extract district codes and mean income

```python
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = pd.read_excel('Appendix-converted.xlsx')
```

```python
df = df.iloc[8:75].drop(range(38,46))
```

```python
df
```

```python
#keep just two of the columns
mean_income2013 = df[['New York City Community District DataTable','Unnamed: 11']].copy()
```

```python
mean_income2013.reset_index(drop = True, inplace = True);
```

```python
mean_income2013.columns[0]
```

```python
#rename the columns
mean_income2013.rename({mean_income2013.columns[0]:'Community_District', mean_income2013.columns[1]:'Mean_Income'}, axis = 1, \
                       inplace = True)
```

```python
#remove star from community_district
mean_income2013['Community_District'] = mean_income2013['Community_District'].str.rstrip('*').copy()
```

```python
#first 5 rows of the new cleaned dataset with just community district and income
mean_income2013.head()
```

```python
#convert file to csv to use in other data analysis
mean_income2013.to_csv('mean_income2013.csv',index= False)
```
