```python
import urllib.request
```

```python
#url for the data I scraped
url = 'https://en.wikipedia.org/wiki/Neighborhoods_in_New_York_City'
```

```python
#put html data in page variable
page = urllib.request.urlopen(url)
```

```python
from bs4 import BeautifulSoup
```

```python
#parase html in beautifulsoup lxml format
soup = BeautifulSoup(page, "lxml")
```

```python
print(soup.prettify())
```

```python
#find the table in our soup
cb_table= soup.find_all('table',class_='wikitable sortable')
```

```python
cb_table[0].findAll('tr')[1].findAll('td')
```

```python
#add the data from the tables into separate lists
cb = []
area = []
pop =[]
pop_km = []
neighborhood = []
```

```python
for row in cb_table[0].findAll('tr')[1:]:
            cb.append(row.findAll('td')[0].text)
            area.append(row.findAll('td')[1].text)
            pop.append(row.findAll('td')[2].text)
            pop_km.append(row.findAll('td')[3].text)
            neighborhood.append(row.findAll('td')[4].text)
```

```python
import pandas as pd
```

```python
#turn the list into a dataframe
df =  pd.DataFrame([cb,area,pop,pop_km,neighborhood]).transpose()
```

```python
#rename column names
df = df.rename({0:'Comm_Boundary',1:'area_Kmsq',2:'Pop_2010_Census',3:'Pop/Kmsq', 4:'Neighborhoods'},axis = 1)
```

```python
df.head()
```

```python
#remove the trailing \n in the Neighborhoods column
df['Neighborhoods'] = df['Neighborhoods'].str.rstrip('\n')
```
